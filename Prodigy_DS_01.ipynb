```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate synthetic datasets
np.random.seed(50)  # For reproducibility
```

```python
# Age distribution
ages = np.random.randint(0, 100, size=1000)  # 1000 random ages between 0 and 100
age_df = pd.DataFrame({'Age': ages})
```

```python
# Gender distribution
genders = np.random.choice(['Male', 'Female', 'Other'], size=1000, p=[0.45, 0.45, 0.1])  # 1000 random genders
gender_df = pd.DataFrame({'Gender': genders})
```

```python
# Plot age distribution (histogram)
plt.figure(figsize=(10, 6))
plt.hist(age_df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.show()
```

```python
# Plot gender distribution (bar chart)
gender_counts = gender_df['Gender'].value_counts()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Distribution of Genders')
plt.show()
```
