```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('/content/train.csv')
```

# Data Cleaning

```python
# Check for missing values
print(df.isnull().sum())

# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df['Cabin'].fillna('Unknown', inplace=True)

# Convert data types
df['Pclass'] = df['Pclass'].astype('category')
```

# **EDA**

```python
# Summary statistics
print(df.describe())
```

```python
# Distribution of Age
sns.histplot(df['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

# Survival rate by Pclass
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Pclass')
plt.show()

# Survival rate by Sex
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.select_dtypes(include=['number']).corr() # Select only numeric columns
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Pairplot (Consider selecting a subset of numeric columns for better visualization)
sns.pairplot(df[['Survived', 'Age', 'Fare']], hue='Survived') # Example with a subset
plt.show()
```
